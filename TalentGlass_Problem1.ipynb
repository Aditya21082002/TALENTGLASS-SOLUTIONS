```python
def find_all_movies(actor_name):
    from selenium import webdriver
    from time import sleep
    from selenium.webdriver.common.by import By
    import pandas as pd
    path = 'C://chromedriver.exe'
    browser = webdriver.Chrome(executable_path=path)
    browser.get('https://www.themoviedb.org/')
    browser.maximize_window()
    # get the input elements
    input_search = browser.find_element('id',"inner_search_v4")
    search_button = browser.find_element('xpath',"//*[@id='inner_search_form']/input")
    # send the input to the webpage
    input_search.send_keys(actor_name)
    sleep(1)
    search_button.click()
    actor_url=browser.find_element(By.CLASS_NAME,"result").get_attribute('href')
    browser.get(actor_url)
    filmography_section = browser.find_elements(By.CLASS_NAME,"credit_group")
    filmography = []
    for film in filmography_section:
       title = film.find_element(By.CLASS_NAME,"tooltip").text
       year = film.find_element(By.CLASS_NAME, "year").text
       filmography.append((year,title))
    # # Print the filmography in descending order of release date
    all_movies_dataset=pd.DataFrame(sorted(filmography, key=lambda x: x[0], reverse=True),columns=["Year","Movie"])
   #  for film in sorted(filmography, key=lambda x: x[1], reverse=True):
   #     print(f"{film[1]} ({film[0]})") 
    all_movies_dataset.drop_duplicates()
    all_movies_dataset.reset_index(drop=True)
    browser.quit()     
    return all_movies_dataset
```

```python
actor_name=input()
all_movies=find_all_movies(actor_name)
```
